```python
import pandas as pd

def load_unidil_data(file):
    df = pd.read_csv(file, header=None)

    # DEBUG: print first few rows
    print("Preview:")
    print(df.head(10))

    # Try different date rows if needed
    date_row = df.iloc[1]

    data = []
    current_section = None

    for i in range(len(df)):
        label = str(df.iloc[i, 0]).strip().lower()

        # 🔥 Flexible section detection
        if "planned" in label:
            current_section = "Planned"
            continue
        elif "actual" in label:
            current_section = "Actual"
            continue
        elif "yield" in label:
            current_section = "Yield"
            continue
        elif "rejection" in label:
            current_section = "Rejection"
            continue
        elif "stoppage" in label:
            current_section = "Stoppage"
            continue

        if current_section:

            # 🔥 Flexible process detection
            if "corrugator" in label:
                process = "Corrugator"
            elif "tuber" in label:
                process = "Tuber"
            elif "printing" in label:
                process = "Printing"
            elif "finishing" in label:
                process = "Finishing"
            else:
                continue

            for col in range(1, len(df.columns)):
                value = df.iloc[i, col]
                date = date_row[col]

                if pd.notna(value) and str(value).strip() not in ["-", ""]:
                    try:
                        data.append({
                            "Date": date,
                            "Process": process,
                            "Metric": current_section,
                            "Value": float(value)
                        })
                    except:
                        pass

    clean_df = pd.DataFrame(data)

    print("Extracted rows:", len(clean_df))  # DEBUG

    if clean_df.empty:
        print("⚠️ No data extracted — check labels or date row")
        return clean_df

    final_df = clean_df.pivot_table(
        index=["Date", "Process"],
        columns="Metric",
        values="Value"
    ).reset_index()

    return final_df
```

```python
df = load_unidil_data("Daily Production Data_NEW Format - UNIDIL.csv")
print(df.head())
```

```python
def add_kpis(df):

    df["Efficiency"] = (df["Actual"] / df["Planned"]) * 100
    df["Production_Loss"] = df["Planned"] - df["Actual"]

    return df
```

```python
def analyze_factory(df):
    insights = []

    for _, row in df.iterrows():

        process = row["Process"]
        date = row["Date"]

        # Efficiency issue
        if row["Efficiency"] < 95:
            insights.append(f"{date} [{process}] low efficiency ({row['Efficiency']:.1f}%)")

        # Production gap
        if row["Actual"] < row["Planned"]:

            if row["Stoppage"] > 60:
                insights.append(f"{date} [{process}] loss due to high downtime")

            elif row["Rejection"] > 1:
                insights.append(f"{date} [{process}] loss due to high rejection")

            else:
                insights.append(f"{date} [{process}] unexplained production loss")

    return insights
```

```python
import matplotlib.pyplot as plt

def plot_trends(df):

    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

    grouped = df.groupby("Date")["Actual"].sum()

    plt.figure()
    grouped.plot()
    plt.title("Daily Production Trend")
    plt.show()
```
